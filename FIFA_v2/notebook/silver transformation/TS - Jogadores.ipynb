```python
import pandas as pd
import json
```

```python
def transform_jogadores():

    # Carregar o arquivo JSON
    with open('/lakehouse/default/Files/raw/jogadores.json', 'r') as f:
        json_list = json.load(f)

    # Criar um DataFrame diretamente a partir da lista de dicionários
    df = pd.DataFrame(json_list)

    # Renomear as colunas
    df.rename(columns={
        'id': 'ID',
        'resourceId': 'ID_Recurso',
        'resourceBaseId': 'ID_Base_Recurso',
        'futBinId': 'ID_FutBin',
        'futWizId': 'ID_FutWiz',
        'club': 'ID_Clube',
        'league': 'ID_Liga',
        'nation': 'ID_Nacionalidade',
        'firstName':'Primeiro_Nome',
        'lastName': 'Sobrenome',
        'name': 'Nome',
        'commonName': 'Nome_Comun',
        'age': 'Idade',
        'birthDate': 'Data_Nascimento',
        'position': 'Posição',
        'positionAlternatives': 'Posição_Alternativa',
        'rating': 'Overall',
        'skillMoves': 'Movimento_Habilidade',
        'rarity': 'Raridade',
        'playStyles': 'Estilo_Jogo',
        'playStylesPlus':'Estilo_Jogo_Plus',
        'foot': 'Pé_Bom',
        'weakFoot': 'Pé_Ruim',
        'attackWorkRate':'Taxa_Trabalho_Ataque',
        'defenseWorkRate':'Taxa_Trabalho_Defesa',
        'totalStats':'Estatistica_Total',
        'totalStatsInGame':'Estatistica_Total_EmJogo',
        'color':'Cor',
        'height': 'Altura',
        'weight': 'Peso',
        'gender': 'Genero',
        'shooting': 'Chute',
        'paceAttributes':'Atributo_Ritmo',
        'ratingAverage':'Classificação_Média',
        'shootingAttributes':'Atributo_Chute',
        'passingAttributes':'Atributo_Passe',
        'dribblingAttributes':'Atributo_Drible',
        'defendingAttributes':'Atributo_Defesa',
        'physicalityAttributes':'Atributo_Fisico',
        'pace': 'Velocidade',
        'passing': 'Passe',
        'dribbling': 'Drible',
        'defending': 'Defesa',
        'physicality': 'Fisico',
        'goalkeeperAttributes': 'Atributo_de_Goleiro'
    }, inplace=True)

    # Preenchimento de valores NaN
    df.fillna({'ID_Recurso':0,'ID_Base_Recurso':0,'ID_FutBin':0,'ID_FutWiz':0,
                'Atributo_de_Goleiro': 0, 'Peso':0,'Idade':0, 'Genero':'-',
                'Movimento_Habilidade':0,'Atributo_Ritmo':0,'Atributo_Chute':0,
                'Atributo_Passe':0,'Atributo_Drible':0,'Atributo_Defesa':0,
                'Atributo_Fisico':0, 'Classificação_Média':0}, inplace=True)

    # Substituir as strings que não correspondem ao formato de data por None
    df['Data_Nascimento'] = pd.to_datetime(df['Data_Nascimento'], errors='coerce')

    # Converter os tipos de dados
    df = df.astype({
        'ID': int,
        'ID_Recurso':int,
        'ID_Base_Recurso':int,
        'ID_FutBin':int,
        'ID_FutWiz':int,
        'ID_Clube': int,
        'ID_Liga': int,
        'ID_Nacionalidade': int,
        'Primeiro_Nome': str,
        'Sobrenome':str,
        'Nome_Comun':str,
        'Nome': str,
        'Idade': int,
        'Data_Nascimento': 'datetime64[ns]',
        'Posição': str,
        'Posição_Alternativa':str,
        'Movimento_Habilidade':int,
        'Raridade':str,
        'Estilo_Jogo':str,
        'Estilo_Jogo_Plus':str,
        'Overall': int,
        'Pé_Bom': str,
        'Pé_Ruim':str,
        'Taxa_Trabalho_Ataque':str,
        'Taxa_Trabalho_Defesa':str,
        'Estatistica_Total':float,
        'Estatistica_Total_EmJogo':float,
        'Cor':str,
        'Altura': float,
        'Peso': int,
        'Genero': str,
        'Chute': int,
        'Atributo_Ritmo':int,
        'Classificação_Média':int,
        'Atributo_Chute':int,
        'Atributo_Passe':int,
        'Atributo_Drible':int,
        'Atributo_Defesa':int,
        'Atributo_Fisico':int,
        'Velocidade': int,
        'Passe': int,
        'Drible': int,
        'Defesa': int,
        'Fisico': int,
        'Atributo_de_Goleiro': int
    })

    df.to_csv('/lakehouse/default/Files/silver/jogadores.csv',sep=',',index=False)

transform_jogadores()
```
