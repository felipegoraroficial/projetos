```python
import pandas as pd
import json
```

```python
def transform_pais():

    with open('/lakehouse/default/Files/raw/pais.json', 'r') as f:
        json_list = json.load(f)

    # Criar um DataFrame diretamente a partir da lista de dicion√°rios
    df = pd.DataFrame(json_list)

    # Renomear as colunas
    df.rename(columns={
        'id': 'ID_Nacionalidade',
        'name': 'Nacionalidade'

    }, inplace=True)

    # Converter os tipos de dados
    df = df.astype({
        'ID_Nacionalidade': int,
        'Nacionalidade': str
    })

    df.to_csv('/lakehouse/default/Files/silver/pais.csv',sep=',',index=False)

transform_pais()
```
