```python
import pandas as pd
import json
```

```python
def transform_liga():


    with open('/lakehouse/default/Files/raw/liga.json', 'r') as f:
        json_list = json.load(f)

    # Criar um DataFrame diretamente a partir da lista de dicion√°rios
    df = pd.DataFrame(json_list)

    # Renomear as colunas
    df.rename(columns={
        'id': 'ID_Liga',
        'name': 'Liga',
        'nationId':'ID_Nacionalidade',
        'gender':'Genero'

    }, inplace=True)

    # Preenchimento de valores NaN
    df.fillna({'ID_Liga':0,'ID_Nacionalidade':0}, inplace=True)

    # Converter os tipos de dados
    df = df.astype({
        'ID_Liga': int,
        'Liga': str,
        'ID_Nacionalidade': int,
        'Genero': str
    })

    df.to_csv('/lakehouse/default/Files/silver/liga.csv',sep=',',index=False)

transform_liga()
```
