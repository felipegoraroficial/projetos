```python
import pandas as pd
import json
```

```python
def transform_clube():

    with open('/lakehouse/default/Files/raw/clubes.json', 'r') as f:
        json_list = json.load(f)

    # Criar um DataFrame diretamente a partir da lista de dicion√°rios
    df = pd.DataFrame(json_list)

    # Renomear as colunas
    df.rename(columns={
        'id': 'ID_Clube',
        'name': 'Clube',
        'league' : 'ID_Liga'

    }, inplace=True)

    # Preenchimento de valores NaN
    df.fillna({'ID_Clube':0,'ID_Liga':0}, inplace=True)

    # Converter os tipos de dados
    df = df.astype({
        'ID_Clube': int,
        'Clube': str,
        'ID_Liga': int
    })

    df.to_csv('/lakehouse/default/Files/silver/clube.csv',sep=',',index=False)

transform_clube()
```
