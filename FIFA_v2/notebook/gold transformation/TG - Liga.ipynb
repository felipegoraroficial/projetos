```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, StringType
```

```python
df = spark.read.format('csv').option('header','true').load('Files/silver/liga.csv')
```

```python
df = df.filter(df.ID_Liga != 0)
df = df.filter(df.ID_Nacionalidade != 0)
```

```python
# Converter os tipos de dados
df = df.withColumn('ID_Liga', col('ID_Liga').cast(IntegerType()))\
    .withColumn('Liga', col('Liga').cast(StringType()))\
    .withColumn('ID_Nacionalidade', col('ID_Nacionalidade').cast(IntegerType()))\
    .withColumn('Genero', col('Genero').cast(StringType()))
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("liga")
```
