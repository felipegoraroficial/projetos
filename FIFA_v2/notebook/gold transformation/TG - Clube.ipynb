```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, StringType
```

```python
df = spark.read.format('csv').option('header','true').load('Files/silver/clube.csv')
```

```python
df = df.filter(df.ID_Liga != 0)
df = df.filter(df.ID_Clube != 0)
```

```python
# Converter os tipos de dados
df = df.withColumn('ID_Clube', col('ID_Clube').cast(IntegerType()))\
    .withColumn('Clube', col('Clube').cast(StringType()))\
    .withColumn('ID_Liga', col('ID_Liga').cast(IntegerType()))
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("clube")
```
