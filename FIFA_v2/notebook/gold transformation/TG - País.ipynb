```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, StringType
```

```python
df = spark.read.format('csv').option('header','true').load('Files/silver/pais.csv')
```

```python
df = df.withColumn('ID_Nacionalidade', col('ID_Nacionalidade').cast(IntegerType()))\
    .withColumn('Nacionalidade', col('Nacionalidade').cast(StringType()))
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("pais")
```
