```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date
from pyspark.sql.types import IntegerType, StringType, FloatType
```

```python
df = spark.read.format('csv').option('header','true').load('Files/silver/jogadores.csv')
```

```python
df = df.filter(df.ID != 0)
df = df.filter(df.ID_Liga != 0)
df = df.filter(df.ID_Clube != 0)
df = df.filter(df.ID_Nacionalidade != 0)
```

```python
# Lista de colunas na tabela SQL
colunas_tabela_sql = ["ID_Clube", "ID_Liga", "ID_Nacionalidade", "Nome", "Idade", "Data_Nascimento",
                      "Posição", "Overall", "Pé_Bom", "Altura", "Peso", "Genero", "Chute", "Velocidade", "Passe",
                      "Drible", "Defesa", "Fisico", "Atributo_de_Goleiro"]

# Filtrar o DataFrame para conter apenas as colunas da tabela SQL
df = df.select(*colunas_tabela_sql)
```

```python
df = df.withColumn('ID_Clube', col('ID_Clube').cast(IntegerType()))\
    .withColumn('ID_Liga', col('ID_Liga').cast(IntegerType()))\
    .withColumn('ID_Nacionalidade', col('ID_Nacionalidade').cast(IntegerType()))\
    .withColumn('Nome', col('Nome').cast(StringType()))\
    .withColumn('Idade', col('Idade').cast(IntegerType()))\
    .withColumn('Data_Nascimento', to_date(col('Data_Nascimento'), 'yyyy-MM-dd'))\
    .withColumn('Posição', col('Posição').cast(StringType()))\
    .withColumn('Overall', col('Overall').cast(IntegerType()))\
    .withColumn('Pé_Bom', col('Pé_Bom').cast(StringType()))\
    .withColumn('Altura', col('Altura').cast(FloatType()))\
    .withColumn('Peso', col('Peso').cast(IntegerType()))\
    .withColumn('Genero', col('Genero').cast(StringType()))\
    .withColumn('Chute', col('Chute').cast(IntegerType()))\
    .withColumn('Velocidade', col('Velocidade').cast(IntegerType()))\
    .withColumn('Passe', col('Passe').cast(IntegerType()))\
    .withColumn('Drible', col('Drible').cast(IntegerType()))\
    .withColumn('Defesa', col('Defesa').cast(IntegerType()))\
    .withColumn('Fisico', col('Fisico').cast(IntegerType()))\
    .withColumn('Atributo_de_Goleiro', col('Atributo_de_Goleiro').cast(IntegerType()))
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("jogadores")
```
