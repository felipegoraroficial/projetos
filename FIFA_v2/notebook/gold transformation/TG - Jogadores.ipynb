```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date
from pyspark.sql.types import IntegerType, StringType, FloatType
```

```python
df = spark.read.format('csv').option('header','true').load('Files/silver/jogadores.csv')
```

```python
df = df.filter(df.ID != 0)
df = df.filter(df.ID_Liga != 0)
df = df.filter(df.ID_Clube != 0)
df = df.filter(df.ID_Nacionalidade != 0)
```

```python
df = df.withColumn('ID', col('ID').cast(IntegerType()))\
    .withColumn('ID_Recurso', col('ID_Recurso').cast(IntegerType()))\
    .withColumn('ID_Base_Recurso', col('ID_Base_Recurso').cast(IntegerType()))\
    .withColumn('ID_FutBin', col('ID_FutBin').cast(IntegerType()))\
    .withColumn('ID_FutWiz', col('ID_FutWiz').cast(IntegerType()))\
    .withColumn('ID_Clube', col('ID_Clube').cast(IntegerType()))\
    .withColumn('ID_Liga', col('ID_Liga').cast(IntegerType()))\
    .withColumn('ID_Nacionalidade', col('ID_Nacionalidade').cast(IntegerType()))\
    .withColumn('Primeiro_Nome', col('Primeiro_Nome').cast(StringType()))\
    .withColumn('Sobrenome', col('Sobrenome').cast(StringType()))\
    .withColumn('Nome_Comun', col('Nome_Comun').cast(StringType()))\
    .withColumn('Nome', col('Nome').cast(StringType()))\
    .withColumn('Idade', col('Idade').cast(IntegerType()))\
    .withColumn('Data_Nascimento', to_date(col('Data_Nascimento'), 'yyyy-MM-dd'))\
    .withColumn('Posição', col('Posição').cast(StringType()))\
    .withColumn('Posição_Alternativa', col('Posição_Alternativa').cast(StringType()))\
    .withColumn('Movimento_Habilidade', col('Movimento_Habilidade').cast(IntegerType()))\
    .withColumn('Raridade', col('Raridade').cast(StringType()))\
    .withColumn('Estilo_Jogo', col('Estilo_Jogo').cast(StringType()))\
    .withColumn('Estilo_Jogo_Plus', col('Estilo_Jogo_Plus').cast(StringType()))\
    .withColumn('Overall', col('Overall').cast(IntegerType()))\
    .withColumn('Pé_Bom', col('Pé_Bom').cast(StringType()))\
    .withColumn('Pé_Ruim', col('Pé_Ruim').cast(StringType()))\
    .withColumn('Taxa_Trabalho_Ataque', col('Taxa_Trabalho_Ataque').cast(StringType()))\
    .withColumn('Taxa_Trabalho_Defesa', col('Taxa_Trabalho_Defesa').cast(StringType()))\
    .withColumn('Estatistica_Total', col('Estatistica_Total').cast(FloatType()))\
    .withColumn('Estatistica_Total_EmJogo', col('Estatistica_Total_EmJogo').cast(FloatType()))\
    .withColumn('Cor', col('Cor').cast(StringType()))\
    .withColumn('Altura', col('Altura').cast(FloatType()))\
    .withColumn('Peso', col('Peso').cast(IntegerType()))\
    .withColumn('Genero', col('Genero').cast(StringType()))\
    .withColumn('Chute', col('Chute').cast(IntegerType()))\
    .withColumn('Atributo_Ritmo', col('Atributo_Ritmo').cast(IntegerType()))\
    .withColumn('Classificação_Média', col('Classificação_Média').cast(IntegerType()))\
    .withColumn('Atributo_Chute', col('Atributo_Chute').cast(IntegerType()))\
    .withColumn('Atributo_Passe', col('Atributo_Passe').cast(IntegerType()))\
    .withColumn('Atributo_Drible', col('Atributo_Drible').cast(IntegerType()))\
    .withColumn('Atributo_Defesa', col('Atributo_Defesa').cast(IntegerType()))\
    .withColumn('Atributo_Fisico', col('Atributo_Fisico').cast(IntegerType()))\
    .withColumn('Velocidade', col('Velocidade').cast(IntegerType()))\
    .withColumn('Passe', col('Passe').cast(IntegerType()))\
    .withColumn('Drible', col('Drible').cast(IntegerType()))\
    .withColumn('Defesa', col('Defesa').cast(IntegerType()))\
    .withColumn('Fisico', col('Fisico').cast(IntegerType()))\
    .withColumn('Atributo_de_Goleiro', col('Atributo_de_Goleiro').cast(IntegerType()))
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("jogadores")
```
