```python
import requests
import json
import pandas as pd



api_key = "7b1b94b8-78ca-4f43-b31c-9b3f4199a1a4" 
headers = {"X-AUTH-TOKEN": api_key}
```

```python
def jogadores():
    jogadores_list  = list()

    for i in range(1, 929):
        url = f"https://futdb.app/api/players?page={i}" 
        response = json.loads(requests.request("GET", url=url, headers=headers).text)

        for item in response['items']:
            jogadores_list.append(item)

    with open('/lakehouse/default/Files/raw/jogadores.json', 'w') as f:
        json.dump(jogadores_list, f)

jogadores()
```
