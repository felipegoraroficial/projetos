```python
import requests
import json
import pandas as pd



api_key = "7b1b94b8-78ca-4f43-b31c-9b3f4199a1a4" 
headers = {"X-AUTH-TOKEN": api_key}
```

```python
def pais():
    nacao_list  = list()

    for i in range(1, 12):
        url = f"https://futdb.app/api/nations?page={i}" 
        response = json.loads(requests.request("GET", url=url, headers=headers).text)

        for item in response['items']:
            nacao_list.append(item)

    with open('/lakehouse/default/Files/raw/pais.json', 'w') as f:
        json.dump(nacao_list, f)

pais()
```
